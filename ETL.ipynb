2. Transform data

2.1 Enrollies data

```python
enrollies_data.info()
```

```python
enrollies_data=enrollies_data.fillna("unknow")
enrollies_data.info()
```

```python
# Chuyn i kiu d liu
enrollies_data['full_name'] = enrollies_data['full_name'].astype('string')
enrollies_data['city'] = enrollies_data['city'].astype('category')
enrollies_data['gender'] = enrollies_data['gender'].astype('category')

# Kim tra li kt qu
enrollies_data.info()
```

2.2 Enrollies' education

```python
enrollies_education.info()
```

```python
enrollies_education=enrollies_education.fillna("unknow")
```

```python
enrollies_education['enrolled_university'] = enrollies_education['enrolled_university'].astype('string')
enrollies_education['education_level'] = enrollies_education['education_level'].astype('category')
enrollies_education['major_discipline'] = enrollies_education['major_discipline'].astype('category')
```

```python
enrollies_education.info()
```

2.3 Enrollies' working experience

```python
enrollies_working_experience.info()
```

```python
enrollies_working_experience= enrollies_working_experience.fillna("unknow")
```

```python
cat_cols = ['relevent_experience','experience','company_size','company_type','last_new_job']
for col in cat_cols:
    enrollies_working_experience[col] = enrollies_working_experience[col].astype('category')
```

```python
enrollies_working_experience.info()
```

```python
df=enrollies_data.copy()
df=df.merge(enrollies_education,on='enrollee_id',how='left')
df=df.merge(enrollies_working_experience,on='enrollee_id',how='left')
df=df.merge(training_hour,on='enrollee_id',how='left')
df=df.merge(employment,on='enrollee_id',how='left')
df=df.merge(city_development_index,on='city',how='left')
df.head()
```

```python
df.to_csv("data.csv",index=False)
```
